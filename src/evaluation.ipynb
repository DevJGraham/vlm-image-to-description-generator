# Evaluation & Categorizing Descriptions

This file takes in the model-generated descriptions, the baseline descriptions, as well as the categories of the items.

A human labeler will randomly and anonymously be given the model and baseline descriptions and evalute the descriptions. The evaluation options are as follows:
- Description 1 better
- Description 2 better
- Both good
- Both bad

The model will than be evaluated for how well it performed, as well as a more granular dive into how well it performed across the different categories of items.

## Merge into one df

```python
both_descriptions = (
  baseline_df[["item_number", "baseline_desc"]]
  .merge(model_df[["item_number", "model_desc"]], on="item_number", how="left")
)
both_descriptions.head()
```

# Randomize the order of the model/baseline descriptions and keep track of the source

```python
def randomize(row):
  # Each option: (column_name, description_text, source_label)
  options = [
    (row["model_desc"], "model"),
    (row["baseline_desc"], "baseline")
  ]
  random.shuffle(options)

  return pd.Series({
    "item_number": row["item_number"],
    "desc1": options[0][0],
    "desc1_source": options[0][1],
    "desc2": options[1][0],
    "desc2_source": options[1][1]
  })
```

```python
test = both_descriptions.apply(randomize, axis=1)
test.head()
```

```python
wrap_width = 80

def human_evaluation(row):
  print(f'Item number {row['item_number']}:')
  print("Description 1:")
  print(textwrap.fill(row["desc1"], width=wrap_width))
  print()

  print("Description 2:")
  print(textwrap.fill(row["desc2"], width=wrap_width))
  print()

  print("\nChoices:")
  print("1 = Description 1 better")
  print("2 = Description 2 better")
  print("3 = Both good")
  print("4 = Both bad")

  while True:
    choice = input("Your choice: ")
    if choice in {"1", "2", "3", "4"}:
      return int(choice)
```

```python
def get_start_point(max_item = 60):
  while True:
    start_point = input(
      f"Choose an item number to start at (1 to {max_item}): "
    )

    # Handle non-integers gracefully
    if not start_point.isdigit():
      print(" Invalid input. Please enter a number.")
      continue

    start_point = int(start_point)

    # Check range
    if 1 <= start_point <= max_item:
      return start_point
    else:
      print(f" Number out of range. Please choose between 1 and {max_item}.")
```

```python
def evaluate_descriptions(df, max_items = 60):
  start_point = get_start_point()

  for item in range(start_point, max_items + 1):
    row = df[df['item_number'] == item].iloc[0]
    choice = human_evaluation(row)
    match choice:
      case 1:
        human_eval = row['desc1_source']
      case 2:
        human_eval = row['desc2_source']
      case 3:
        human_eval = 'Both descriptions are acceptable'
      case 4:
        human_eval = 'Both descriptions are unnacceptable'
      case _:
        human_eval = f'Error recording response for item {item}'

    df.loc[df['item_number'] == item, 'human_label'] = human_eval

    clear_output(wait=False)
    sleep(0.5)

  print(f' Finished evaluating items')
```

# Final Evaluation

```python
df1 = pd.read_csv('Aypt2025-13_evaluated.csv')
df2 = pd.read_csv('auction_13_categories.csv', index_col=0)

eval = pd.merge(df1, df2.drop(columns=['Description'], axis=1), how='left', on='item_number')
```

```python
eval.head()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x=eval['human_label'])

plt.title("Human Evaluation Results")
plt.xlabel("Human Label")
plt.ylabel("# of Descriptions")

for c in ax.containers:
  ax.bar_label(c)

plt.show()
```

```python
label_counts = eval['human_label'].value_counts()
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%')

plt.title("Human Evaluation Results (%)")

plt.show()
```

```python
infrequent_indexes = eval['category'].value_counts()[lambda x: x <= 2].index
eval['category_grouped'] = eval['category'].replace(infrequent_indexes, 'Other')
```

```python
category_counts = eval['category_grouped'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')

plt.title("Percentages of Categories Among Descriptions")

plt.show()
```

```python
category_labels = eval.groupby('category_grouped')['human_label'].value_counts().reset_index()
```

```python
plt.figure(figsize=(13, 5))
ax = sns.barplot(data=category_labels, x='category_grouped', y='count', hue='human_label')

plt.title("Model Performance Across Categories")
plt.xlabel("Description Category")
plt.ylabel("# of Descriptions")

for c in ax.containers:
  ax.bar_label(c)

plt.show()
```

```python
# Create a df that contains the performance of across the different categories
category_perf = (
    category_labels.groupby('category_grouped')[['count', 'human_label']]
    .apply(lambda x: sum(
        x[x['human_label'].str.contains('Both descriptions are acceptable|model')]['count']
    ) / sum(x['count']))
    .reset_index(name='success_rate')
)

# Create a df that contains the count of each category
category_counts = (
    category_labels.groupby('category_grouped')['count']
    .sum()
    .reset_index(name='total_count')
)

# Merge the two
category_perf = category_perf.merge(category_counts, on='category_grouped')

# Create a simple weighted score based on the performance and the count of each category
category_perf['weighted_score'] = category_perf['success_rate'] * category_perf['total_count']

# Reformat the success rate
category_perf['success_rate'] = category_perf['success_rate'].apply(lambda x: f"{round(x * 100, 1)}%")
```

```python
category_perf.sort_values('weighted_score', ascending=-False)
```

## Conclusion

The model demonstrates promising potential for improving auction description writing workflows, though there is still room for refinement.

Out of 60 evaluated descriptions, 25 (41.7%) of the model-generated outputs were rated as either comparable to or better than the baseline descriptions.

The most frequent item categories in the dataset were *Tools* (28.3%), *Other* (21.7%), and *Art* (15%).

The strongest model performance was observed in the Tools category, with a success rate of 47.1% (defined as cases where the models description was rated equally or better than the baseline) and a weighted score of 8 based on its frequency.

In summary, the model can reliably generate initial draft descriptions that serve as a solid starting point. Approximately 41% of the outputs can be used as-is, while the remaining 59% will require light human revision to ensure accuracy and completeness. This represents a meaningful reduction in manual workload, with clear potential for further improvement as the model continues to evolve.
